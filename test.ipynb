```python

from DLS import DLS_class
import matplotlib.pyplot as plt
import numpy as np

DLS=DLS_class(dirname=r'/Users/new_admin/nextcloud/DLS/Ferritin/stockdil8_50%wgly_temp_dep')
DLS.load_data()
```

```python

DLS.average_data()
```

```python


popt2,q,D2,temperatures=DLS.plot_fit_g2(function=DLS.exponential,tnorm=10)
```

```python
plt.figure
plt.plot(1000/np.array(temperatures),D2,marker='o')

plt.grid('on')

plt.yscale('log')
plt.xlabel('1000/T (1/K)')
plt.ylabel('D (m$^2$/s)')
```
